```python
def push(a, x):
    a.append(x)
    return a

def pop(a):
    if len(a) == 0:
        print('Stack is already empty')
    else:
        a.pop()
        return a

def peek(a):
    if len(a) == 0:
        print('Stack is already empty')
    else:
        print(a[len(a) - 1])

def print_stk(a):
    if len(a) == 0:
        print('Stack is already empty')
    else:
        for i in range(len(a)):
            print(a[i], end = " ")
        print()

ch = 0
stk = []

while(ch != 5):
    print('Enter :')
    print('1. To Push an element into the stack')
    print('2. To Pop an element from the stack')
    print('3. To Peek')
    print('4. To Print all the elements in the stack')
    print('5. To Exit')
    
    ch = int(input())
    
    if ch == 1:
        val = int(input('Enter the value to be pushed into the stack : '))
        stk = push(stk, val)
    if ch == 2:
        stk = pop(stk)
    if ch == 3:
        peek(stk)
    if ch == 4:
        print_stk(stk)
```

```python
def enqueue(a, x):
    a.append(x)
    return a

def dequeue(a):
    if len(a) == 0:
        print('Queue is already empty')
    else:
        a.pop(0)
        return a

def peek(a):
    if len(a) == 0:
        print('Queue is already empty')
    else:
        print(a[0])

def print_que(a):
    if len(a) == 0:
        print('Queue is already empty')
    else:
        for i in range(len(a)):
            print(a[i], end = " ")
        print()

ch = 0
que = []

while(ch != 5):
    print('Enter :')
    print('1. To Enqueue an element into the queue')
    print('2. To Dequeue an element from the queue')
    print('3. To Peek')
    print('4. To Print all the elements in the queue')
    print('5. To Exit')
    
    ch = int(input())
    
    if ch == 1:
        val = int(input('Enter the value to be inserted into the queue : '))
        que = enqueue(que, val)
    if ch == 2:
        que = dequeue(que)
    if ch == 3:
        peek(que)
    if ch == 4:
        print_que(que)
```

```python
para = input('Enter a paragraph (min. 3 sentences) : ')

while len(para) < 50:
    para = input('Enter a paragraph (min. 3 sentences) : ')
    print('Length of the entered paragraph is : ',len(para))
    
words = 0
sentences = 0

ch = para[0]
if ch.islower():
    ch = ch.upper()
    para = ch + para[1:]
    
for i in range(len(para)):
    if para[i] == " ":
        words += 1
    if para[i] == ".":
        sentences += 1
        if i+1 < len(para) and para[i + 1].islower():
            ch = para[i+1]
            ch = ch.upper()
            para = para[0:i+1] + ch + para[i+2:]
    if para[i] == 'a' or para[i] == 'A':
        para = para.replace(para[i], 'b')
    if para[i] == 'e' or para[i] == 'E':
        para = para.replace(para[i], 'f')
    if para[i] == 'i' or para[i] == 'I':
        para = para.replace(para[i], 'j')
    if para[i] == 'o' or para[i] == 'O':
        para = para.replace(para[i], 'p')
    if para[i] == 'u' or para[i] == 'U':
        para = para.replace(para[i], 'v')
    if para[i] == 's' or para[i] == 'S':
        para = para.replace(para[i], '$')
        
print('Words : 'words + sentences)
print('Senetences : 'sentences)
print(para)
```
