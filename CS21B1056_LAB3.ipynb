```python
# Question 1

students = {}
ch = 0
while(ch != 4):
    print("Enter : ")
    print("1.Insert")
    print("2.Delete")
    print("3.Search")
    print("4.Exit")

    ch = int(input())

    a = []
    if(ch == 1):
        roll = int(input("Enter Roll number of the student:"))
        
        name = input("Enter name of the student:")
        cgpa = float(input("Enter cgpa of the student:"))
        mobile = int(input("Enter Mobile number of the student:"))
        a.append(name)
        a.append(cgpa)
        a.append(mobile)
        
        students[roll] = a
        print("List inserted successfully")
    elif(ch == 2):
        roll = int(input("Enter roll number of the student you want to delete : "))
        
        if roll in students.keys():
            del students[roll]
            print("Entry Deleted")
        else:
            print("Roll number not found")
    elif(ch == 3):
        roll=int(input("Enter roll number of the student you want to search : "))
        
        if roll in students.keys():
            print(students[roll])
        else:
            print("Roll number not found")
    elif(ch == 4):
        print("Exited successfully")
    else:
        print("Invalid choice")
```

```python
# Question 2

Q = []
P = []

print("Enter quantity and respective price of the order : ")

for i in range(5):
    q = int(input(f"Enter Quantity of {i+1}th item"))
    p = int(input(f"Enter Price of {i+1}th item"))
    Q.append(q)
    P.append(p)
    
print(Q)
print(P)

TP = []
c = 0

for i in range(5):
    k = Q[i]*P[i]
    c = c+k
    TP.append(k)
    
tuple(TP)
    
print("The total price list of above quantity and prices is: ",TP)

print(f"Total amount purchased by all 5 users is : {c}")
```

```python
# Question 3

n = int(input("Enter no.of items : "))
a = []

for i in range(n):
    y = input("Enter name of the item : ")
    x = int(input(f"Enter price of {i+1}th item : "))
    a.append((y,x))
    
a.sort(key=lambda z:z[1])
print(a)
```
